```python
import optuna
import lightgbm as lgb
import numpy as np

def objective(trial, train_dataset, n_folds=5):
    """
    Функция для оптимизации гиперпараметров с использованием кросс-валидации.
    
    Parameters:
    -----------
    trial : optuna.Trial
        Объект trial от Optuna
    train_dataset : lgb.Dataset
        Полный тренировочный датасет
    n_folds : int
        Количество фолдов для кросс-валидации
    
    Returns:
    --------
    float : среднее значение метрики на валидационных фолдах
    """
    params = {
        'objective': 'binary',
        'metric': 'binary_logloss',  # Убедитесь, что эта метрика вам подходит
        'boosting_type': 'gbdt',
        'is_unbalance': True,
        'num_leaves': trial.suggest_int('num_leaves', 20, 300),
        'learning_rate': trial.suggest_float('learning_rate', 0.01, 0.3, log=True),
        'min_child_samples': trial.suggest_int('min_child_samples', 5, 100),
        'subsample': trial.suggest_float('subsample', 0.5, 1.0),
        'colsample_bytree': trial.suggest_float('colsample_bytree', 0.5, 1.0),
        'reg_alpha': trial.suggest_float('reg_alpha', 1e-8, 10.0, log=True),
        'reg_lambda': trial.suggest_float('reg_lambda', 1e-8, 10.0, log=True),
        'max_depth': trial.suggest_int('max_depth', 1, 15),
        'min_split_gain': trial.suggest_float('min_split_gain', 0.0, 0.5),
        'n_estimators': trial.suggest_int('n_estimators', 10, 3000),
        'random_state': 42,  # Важно для воспроизводимости
        'verbosity': -1,  # Отключаем логи
        'n_jobs': -1,  # Используем все ядра
    }
    
    try:
        # Выполняем кросс-валидацию
        cv_results = lgb.cv(
            params,
            train_dataset,
            num_boost_round=1000,
            nfold=n_folds,
            stratified=True,  # Сохраняем распределение классов
            # early_stopping_rounds=50,
            seed=42,
            return_cvbooster=False,  # Не возвращаем все модели
            # verbose_eval=False,  # Не выводим логи
        )
        
        # Получаем лучшую метрику
        # Ключ зависит от метрики, обычно 'valid binary_logloss-mean'
        metric_key = 'valid binary_logloss-mean'
        if metric_key not in cv_results:
            # Пробуем найти ключ с метрикой
            for key in cv_results.keys():
                if 'valid' in key and 'mean' in key:
                    metric_key = key
                    break
        
        best_score = np.min(cv_results[metric_key])
        return best_score
    
    except Exception as e:
        # В случае ошибки возвращаем плохой результат
        print(f"Ошибка в trial {trial.number}: {e}")
        return float('inf')


# Создаем study и запускаем оптимизацию
study = optuna.create_study(
    direction='minimize',  # Минимизируем logloss
    study_name='lgbm_hyperopt',
    storage='sqlite:///lgbm_optuna.db',  # Опционально: сохраняем в базу
    load_if_exists=True,  # Продолжить с предыдущих trials
)

# Запускаем оптимизацию, передавая train_dataset как аргумент
study.optimize(
    lambda trial: objective(trial, train_data, n_folds=5),
    n_trials=50,
    n_jobs=1,  # Для параллельных trials можно увеличить, но будьте осторожны с памятью
    gc_after_trial=True,  # Очистка памяти после каждого trial
    show_progress_bar=True  # Прогресс-бар
)

# Результаты
print("=" * 50)
print(f"Лучшее значение метрики: {study.best_value:.5f}")
print(f"Лучшие параметры:")
for key, value in study.best_params.items():
    print(f"  {key}: {value}")

# Визуализация результатов (опционально)
try:
    import matplotlib.pyplot as plt
    
    # 1. История оптимизации
    fig = plt.figure(figsize=(12, 8))
    
    ax1 = fig.add_subplot(2, 2, 1)
    ax1.plot(range(1, len(study.trials) + 1), [t.value for t in study.trials])
    ax1.set_xlabel('Номер trial')
    ax1.set_ylabel('Значение метрики')
    ax1.set_title('История оптимизации')
    ax1.grid(True)
    
    # 2. Важность гиперпараметров
    ax2 = fig.add_subplot(2, 2, 2)
    optuna.visualization.matplotlib.plot_param_importances(study, ax=ax2)
    ax2.set_title('Важность гиперпараметров')
    
    # 3. Параллельные координаты
    ax3 = fig.add_subplot(2, 2, (3, 4))
    optuna.visualization.matplotlib.plot_parallel_coordinate(study, ax=ax3)
    ax3.set_title('Параллельные координаты')
    
    plt.tight_layout()
    plt.show()
    
except ImportError:
    print("Для визуализации установите matplotlib: pip install matplotlib")

# Обучение финальной модели с лучшими параметрами
print("\n" + "=" * 50)
print("Обучение финальной модели с лучшими параметрами...")

# Добавляем обязательные параметры, которые не оптимизировались
best_params = study.best_params.copy()
best_params.update({
    'objective': 'binary',
    'metric': 'binary_logloss',
    'boosting_type': 'gbdt',
    'random_state': 42,
    'verbosity': 1,
    'n_jobs': -1,
})

# # Создаем валидационный набор для early stopping
# train_data_subset = lgb.Dataset(
#     X_train, 
#     label=y_train,
#     categorical_feature=category_cols if 'category_cols' in locals() else None
# )
# valid_data = lgb.Dataset(
#     X_test, 
#     label=y_test,
#     reference=train_data_subset,
#     categorical_feature=category_cols if 'category_cols' in locals() else None
# )

# # Обучаем финальную модель
# final_model = lgb.train(
#     best_params,
#     train_data_subset,
#     valid_sets=[valid_data],
#     valid_names=['valid'],
#     num_boost_round=1000,
#     early_stopping_rounds=50,
#     verbose_eval=100,
# )

# print(f"Финальная модель обучена! Лучшая итерация: {final_model.best_iteration}")
```
